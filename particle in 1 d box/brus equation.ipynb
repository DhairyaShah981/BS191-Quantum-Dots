```python
# brus equations
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
Eg_bulk1 = 2.4*1.60218e-19
Eg_bulk2 = 2*1.60218e-19
h=6.62607015e-34 #plank's constant
e=1.602176634e-19 #electronic charge
epsilon_0=8.854187817e-12 #epsilon not
pi=3.141592653589793238
c=299792458 #speed of light(metre/sec)
```

```python
# CdS constants
Me1=0.31*9.10938356e-31 #effective mass of electron
Mh1=0.15*9.10938356e-31 #effective mass of hole

Me2=0.26*9.10938356e-31 #effective mass of electron
Mh2=0.19*9.10938356e-31 #effective mass of hole

Me3=0.49*9.10938356e-31 #effective mass of electron
Mh3=0.35*9.10938356e-31 #effective mass of hole

epsilon=6.23 #epsilon
```

### band gap energy dependence on crystal size


```python
def E1(d, Me, Mh):
    return (h**2/(2*d**2))*(Me**(-1)+Mh**(-1))
```

```python
def E(d, Me, Mh, Eg_bulk):
    return Eg_bulk+E1(d, Me, Mh)
```

```python
def E2(d):
    return -((0.9*e**2)/(pi*epsilon*epsilon_0*d))
```

```python
def lambdaa(d):
    return (h*c)/E(d)
```

```python
x_axis = np.array(d1)
print(type(d1[0]))
f = plt.figure()

f.set_figwidth(25)
f.set_figheight(13)
y_axis1 = np.array([E(i*10**(-9), Me1, Mh1, Eg_bulk1) for i in d1])
y_axis2 = np.array([E(i*10**(-9), Me2, Mh2, Eg_bulk2) for i in d1])
y_axis3 = np.array([E(i*10**(-9), Me3, Mh3, Eg_bulk2) for i in d1])
plt.rcParams.update({'font.size': 30})
plt.title('E vs d', fontsize=60)
plt.ylabel('Energy Band Gap (eV)', fontsize=40)
plt.xlabel('Diameter (nm)', fontsize=40)
plt.plot(x_axis,y_axis1)
plt.plot(x_axis,y_axis2)
plt.plot(x_axis,y_axis3)
plt.legend(['Me=0.31 Mh=0.15','Me=0.26 Mh=0.19', 'Me=0.49 Mh=0.35'], loc='upper right')
plt.show()
```
