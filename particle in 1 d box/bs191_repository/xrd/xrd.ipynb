```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open("xrd.txt","r")
f.readline().strip()
print(f.readline())
print(f.readline())
two_theta = []
intensity = []
two_theta_2 = []
intensity_2 = []
for i in range(1,6000   ):
    v = f.readline()
    print(v)
    two_theta.append(float(v.split()[0]))
    intensity.append(float(v.split()[1]))
    two_theta_2.append(float(v.split()[2]))
    intensity_2.append(float(v.split()[3]))
```

```python
f = plt.figure()
plt.rcParams.update({'font.size': 28})
f.set_figwidth(40)
f.set_figheight(13)
plt.xlabel('2 theta', fontsize=40)
plt.ylabel('Intensity', fontsize=40)
plt.plot(two_theta,intensity)
plt.show()
plt.savefig('first_two_columns.png')
```

```python
f = plt.figure()
plt.rcParams.update({'font.size': 28})
f.set_figwidth(40)
f.set_figheight(13)
plt.xlabel('2 theta', fontsize=40)
plt.ylabel('Intensity', fontsize=40)
plt.plot(two_theta_2,intensity_2)
plt.show()
plt.savefig('three_four_columns.png')
```
